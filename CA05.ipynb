## Problem Statement

The objective of this assignment is to build a movie recommendation system using the k-Nearest Neighbors (kNN) algorithm.

Given a dataset of movies with features such as IMDB rating and genre indicators, the goal is to identify the 5 movies that are most similar to a given query movie ("The Post").

This is a similarity-based problem rather than a classification or regression task. The kNN algorithm will be used to compute distances between movies based on their feature vectors and return the closest matches.

#Step 1

Start with gathering the data

```python
import pandas as pd

url = "https://github.com/ArinB/MSBA-CA-Data/raw/main/CA05/movies_recommendation_data.csv"
df = pd.read_csv(url)

df.head()
```

#Step 2 Data Exploration
Look into the data and find patterns and the quality of the data

```python
df.shape
df.info()
df.describe()
```

##Insights
- The "label" column is ignored as it is not relevant for this similarity-based problem.

- All features are numeric, which allows direct application of distance-based algorithms like kNN.

#Step 3 Data Cleaning
Standardize the data to ensure that all variables contribute equally to the distance calculation. This is important because kNN is sensitive to the scale of the data

Drop non-feature columns such as Movie ID, Movie Name, and Label.

```python
from sklearn.preprocessing import StandardScaler

# Drop non-feature columns
X = df.drop(columns=["Movie ID", "Movie Name", "Label"])

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

#Step 4 Building the kNN model

Use the `NearestNeighbors` model to find the 5 nearest neighbors using Euclidean distance. Unlike classification tasks, this model is used purely for similarity search.


```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(n_neighbors=5, metric='euclidean')
model.fit(X_scaled)
```

#Step 5 Create Query Vector ("The Post")

- A feature vector is manually created for the movie "The Post" based on the given attributes.

- This vector is then scaled using the same scaler applied to the dataset to maintain consistency in distance calculations.

```python
import numpy as np

query = np.array([[7.2, 1, 1, 0, 0, 0, 0, 1]])

# Scale query
query_scaled = scaler.transform(query)
```

#Step 6 Get Recommendations

- The kNN model is used to compute the distances between the query movie and all movies in the dataset.

- The 5 closest movies are identified and returned as recommendations.

```python
distances, indices = model.kneighbors(query_scaled)

recommended_movies = df.iloc[indices[0]]

recommended_movies
```
